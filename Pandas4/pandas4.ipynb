## Problem 2

The file *paychecks.csv* contains values of an hourly employee's last 93 paychecks.
Paychecks are given on the first and third Fridays of each month, and the employee started working on March 13, 2008.
Read in the data, using *pd.date_range()* to generate the DatetimeIndex.
Plot the data.
(Hint: use the *union()* method of *DatetimeIndex* class.)

```python
#No header in csv file so header "Salary" was added
pay = pd.read_csv("paychecks.csv",names=["Salary"])

#Date series of third fridays of the month starting 3/13/2008
third = pd.date_range(start= '3/13/2008', periods = 47, freq = 'WOM-3FRI')

#Date series of first fridays of the month starting 3/13/2008
first = pd.date_range(start= '3/13/2008', periods = 46, freq = 'WOM-1FRI')

#Union of the two serieses
date = third.union(first)

#Unioned dates used as the index of the dataframe and then data was plotted
pay["Date Given"] = date
pay.set_index("Date Given").plot(figsize=(9,5), title="Paychecks since March 13, 2008", legend= False)

#Labels
plt.xlabel("Date when paycheck was given")
plt.ylabel("Salary")
plt.show()
```

## Problem 3

The file *finances.csv* contains a list of simulated quarterly earnings and expense totals from a fictional company.
Load the data into a Series or DataFrame with a *PeriodIndex* with a quarterly frequency.
Assume the fiscal year starts at the beginning of September and that the data begins in September 1978.
Plot the data.

```python
fin = pd.read_csv("finances.csv")

#Quarters started on 1978/09 with the periods equal to the number of datapoints
p1 = pd.period_range(start="1978-9", periods=len(fin), freq="Q-AUG")

#The set periods were made the index of the fin dataframe 
fin["Period"]=p1
fin=fin.set_index("Period")

#Plot of earnings and expenses
fin.plot(title="Quarterly Earnings and Expense Totals", figsize = (9,5))
plt.xlabel("Quarter Period")
plt.ylabel("Total Earnings/Expenses")
plt.show()
```

## Problem 4

The file *website_traffic.csv* contains records for different visits to a fictitious website.
Read in the data, calculate the duration of each visit in seconds and convert the index to a *DatetimeIndex*.
Use downsampling to calculate the average visit duration by minute and the average visit duration by hour.
Plot both results on the same graph.

```python
web = pd.read_csv("website_traffic.csv")
#Convert ENTER and LEAVE columns as DatetimeIndex
web.ENTER = pd.to_datetime(web.ENTER)
web.LEAVE = pd.to_datetime(web.LEAVE)

#Compute for duration of visit and converting it to seconds using list comprehension
web["Visit"] = pd.Series([x.seconds for x in (web["LEAVE"] - web["ENTER"])])

#Smaller DataFrame only keeping ENTER (as index) and Visit (duration of visit in seconds)
web1 = web.loc[:,["ENTER","Visit"]].set_index("ENTER")

#Mean of visit duration per minute
mins=web1.resample("min").mean()
mins.columns = ['Average visit duration per minute']

#Mean of visit duration per hour
hr=web1.resample("h").mean()
hr.columns = ['Average visit duration per hour']

#Add 1st hour in minutes df; NaN values left as is
add = mins.loc[datetime(2016,10,7,2)]
add.name = datetime(2016,10,7,1)
minute=mins.append(add).sort_values("ENTER")

#Reindexing hour column to match mins
hour = hr.reindex(minute.index).ffill()

#Plot two dataframes in one subplot
fig1 = plt.figure(figsize=(20,6))
ax1 = fig1.add_subplot(111)

ax1.plot(minute['Average visit duration per minute'], color='lightgray')
ax1.plot(hour['Average visit duration per hour'], color="r")
plt.legend(loc='upper left')
plt.title("Website Traffic on 10/7/2016")
plt.xlabel("Date and Time")
plt.show()
```

## Problem 5
Compute the following information about the DJIA dataset from Problem 1.
- The single day with the largest gain.
- The single day with the largest loss.
- The month with the largest gain.
- The month with the largest loss.

For the monthly statistics, define the gain (or loss) to be the difference between the DJIA on the last and first days of the month.

```python
#Calculate Gain/Loss per day
DJIA["Gain or Loss"] = (DJIA - DJIA.shift(1))

#Sliced DataFrame of DJIA data of gain/loss per day (max, min)
day = pd.DataFrame(DJIA.dropna().sort_values("Gain or Loss",ascending=False).iloc[[0,-1],1])

#Get the first and last days of the month (DataFrame)
djia_m = (DJIA.resample("M").last() - DJIA.resample("M").first()).VALUE
djia_m.index = pd.period_range(start='9/2006',  periods=len(DJIA.resample("M")), freq = 'M')
djia_m = pd.DataFrame(djia_m)
djia_m.index.names = ["DATE"]
djia_m.columns = ['Gain or Loss']
month = djia_m.sort_values("Gain or Loss",ascending=False).iloc[[0,-1],:]

#Summarized largest gain/loss
summary = pd.concat([day, month])
summary[" "] = ["DAY - Largest gain","DAY - Largest loss","MONTH - Largest gain", "MONTH - Largest Loss"]
summary=summary.reset_index().set_index(' ')

summary
```

The **day** with the **largest gain is 10/13/2008** and the one with the **largest loss is 09/29/2008**.

The **month** with the **largest gain is 10/2015** and the day with the **largest loss is 10/2008**.

## Problem 6

Plot the following from the DJIA dataset with a window or span of 30, 120, and 365.
- The original data points.
- Rolling average.
- Exponential average.
- Minimum rolling values.
- Maximum rolling values.

Describe how varying the length of the window changes the approximation to the data.

```python
#Redefine djia dataset (refresh)
djia = pd.read_csv("DJIA.csv")
djia.DATE = pd.to_datetime(djia.DATE, format='%Y-%m-%d')
djia= djia.set_index("DATE")
djia = pd.DataFrame(pd.to_numeric(djia.VALUE, errors= "coerce", downcast='float').dropna())

#Rolling Average of DJIA (30, 100, 365)
ax1 = plt.subplot(121)
djia.plot(color = "gray", lw = 0.3, ax = ax1, figsize = (30,5))
djia.rolling(window = 30).mean().plot(color='r', lw=1, ax=ax1)
djia.rolling(window = 100).mean().plot(color='darkorange', lw=1, ax=ax1)
djia.rolling(window = 365).mean().plot(color='gold', lw=1, ax=ax1)
ax1.legend(["Actual","Rolling (30)","Rolling (100)", "Rolling (365)"], loc="lower right")
ax1.set_title("Rolling Average")
plt.show()

#Exponentially-Weighted Moving Average (EWMA) of DJIA (30, 100, 365)
ax2 = plt.subplot(121)
djia.plot(color = "gray", lw = 0.3, ax = ax2, figsize = (30,5))
djia.ewm(span=30).mean().plot(color='c', lw=1, ax=ax2)
djia.ewm(span=100).mean().plot(color='deepskyblue', lw=1, ax=ax2)
djia.ewm(span=100).mean().plot(color='darkslateblue', lw=1, ax=ax2)
djia.ewm(span=365).mean().plot(color='darkgreen', lw=1, ax=ax2)
ax2.legend(["Actual","EWMA (30)","EWMA (100)", "EWMA (365)"], loc="lower right")
ax2.set_title("Exponentially-Weighted Moving Average (EWMA)")
plt.show()

#Rolling Minimum of DJIA (30, 100, 365)
ax3 = plt.subplot(121)
djia.plot(color = "gray", lw = 0.3, ax = ax3, figsize = (30,5))
djia.rolling(window = 30).min().plot(color='gold', lw=1, ax=ax3)
djia.rolling(window = 100).min().plot(color='plum', lw=1, ax=ax3)
djia.rolling(window = 365).min().plot(color='rebeccapurple', lw=1, ax=ax3)
ax3.legend(["Actual","Rolling (30)","Rolling (100)", "Rolling (365)"], loc="lower right")
ax3.set_title("Rolling Minimum")
plt.show()

#Rolling Maximum of DJIA (30, 100, 365)
ax4 = plt.subplot(121)
djia.plot(color = "gray", lw = 0.3, ax = ax4, figsize = (30,5))
djia.rolling(window = 30).max().plot(color='gold', lw=1, ax=ax4)
djia.rolling(window = 100).max().plot(color='lightseagreen', lw=1, ax=ax4)
djia.rolling(window = 365).max().plot(color='indigo', lw=1, ax=ax4)
ax4.legend(["Actual","Rolling (30)","Rolling (100)", "Rolling (365)"], loc="lower right")
ax4.set_title("Rolling Minimum")
plt.show()
```

From the 4 graphs above, we can see that as increasing the window/span indicated for the rolling/EWM function, the better it is for predicting future values but you sacrifice accuracy since it also gives a less accurate estimate on the data.
